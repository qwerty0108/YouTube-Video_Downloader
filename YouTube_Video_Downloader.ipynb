# PLAYLIST DOWNLOAD WITH PROGRESS BAR

```python
import time
from pytube import Playlist
from pytube import YouTube

def progress(chunk, file_handle, bytes_remaining):
    global filesize
    remaining = (100 * bytes_remaining) / filesize
    step = 100 - int(remaining)
    print("Completed:", str(step)+"%")


# Mahabharat Katha
# p = Playlist("https://www.youtube.com/playlist?list=PLFr_jkwUp0hjr2TcD5fOCqhPpPw78k0_t")

# Mahabharat
p = Playlist("https://www.youtube.com/playlist?list=PLFr_jkwUp0hhm1lR1TSdgESOfoyLQR3t2")

total = 0
print("Total number of videos : ", len(p), "\n")
for num, i in enumerate(p):
    yt = YouTube(i, on_progress_callback = progress)
    for j in yt.streams.filter(progressive=True, res="360p"):
        filesize = j.filesize
        total += filesize
#         print(filesize)
        j.download("./Mahabharat")
        print(str(num+1)+" Done\n")
    time.sleep(1.2)
print("\nTotal : ", total)





```
